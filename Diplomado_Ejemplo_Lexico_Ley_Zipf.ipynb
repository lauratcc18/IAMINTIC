```python
# https://wapp.registraduria.gov.co/electoral/Elecciones-2019/docs/proGobierno/ALC/16/PLAN_AL16001000376_E6.pdf
from nltk.corpus import PlaintextCorpusReader
corpus_root = '/content/planes' # '/content/drive/MyDrive/Dataset/corpora' #'/usr/share/dict'
wordlists = PlaintextCorpusReader(corpus_root, '.*')
print(wordlists.fileids())
print(wordlists.words()[500:550])
```

```python
import nltk
words = wordlists.words("2023_PLAN_GALAN_BOGOTA.txt")
print(words[:50])
text1 = nltk.Text(words)
text1.concordance('seguridad')
```

```python
text1.similar('seguridad')
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (20, 7))
# words = conll2007.words()
# words = word_tokenize(text)
print(len(words))
print(' '.join(words[:100]))
fd = FreqDist (words)
fd.plot(100)
```

```python
fd.tabulate(50)
```

```python
fd = {k: v for k, v in sorted (fd.items (), key = lambda item: item [1], reverse = True)}
```

```python
ranks = list ()
freqs = list ()

for rank, word in enumerate (fd):
      ranks.append (rank + 1)
      freqs.append (fd [word])
```

```python
#Ley de Zipf
fig = plt.figure(figsize = (10, 7))
plt.loglog (ranks, freqs)
plt.xlabel ('Rank')
plt.ylabel ('Freq')
plt.title ('Log - Log rank - freq chart')
plt.show ()
```

```python
from nltk.corpus import  stopwords
nltk.download('stopwords')
list_stop_words_es = stopwords.words('spanish')
```

```python
list_stop_words_es[:50]
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (20, 7))
# words = conll2007.words()
words_2 = [w.lower() for w in words if w.lower() not in list_stop_words_es and w.isalpha()]
fd = FreqDist (words_2)
fd.plot(100)
```

```python
import es_core_news_sm
nlp = es_core_news_sm.load()
doc = nlp(wordlists.raw("2023_PLAN_GALAN_BOGOTA.txt"))
words = [word.lemma_.lower() for word in doc if word.text.isalpha() and word.text.lower() not in list_stop_words_es]
fig = plt.figure(figsize = (20, 7))
# words = conll2007.words()
# words = [w.lower() for w in word_tokenize(text) if w.lower() not in list_stop_words_es and w.isalpha()]
fd = FreqDist (words)
fd.plot(100)
```

```python
fd.tabulate(50)
```

```python
words = [word.lemma_ for word in doc if word.text.isalpha() and word.text.lower() not in list_stop_words_es and word.pos_ == 'ADJ']
fig = plt.figure(figsize = (20, 7))
# words = conll2007.words()
# words = [w.lower() for w in word_tokenize(text) if w.lower() not in list_stop_words_es and w.isalpha()]
fd = FreqDist (words)
fd.plot(100)
```

```python
from nltk.stem import SnowballStemmer

sb = SnowballStemmer('spanish')
#words = [word.lemma_.lower() for word in doc if word.text.isalpha()]
fig = plt.figure(figsize = (20, 7))
# words = conll2007.words()
words = [sb.stem(w.lower()) for w in wordlists.words("2023_PLAN_GALAN_BOGOTA.txt") if w.lower() not in list_stop_words_es and w.isalpha()]
fd = FreqDist (words)
fd.plot(100)
```
