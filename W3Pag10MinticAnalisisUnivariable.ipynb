```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#Cargar datos
url="https://www.datos.gov.co/resource/w9zh-vetq.csv"
#Conjunto de datos
df=pd.read_csv(url)
#obtener estadisticas descriptivas
print(df.describe())
#visualizar histograma y diagrama de caja
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
print(df.columns)
sns.histplot(df['montos_desembolsados'],bins=10,kde=True,color='skyblue')
plt.title('HISTOGRAMA montos_desembolsados')

plt.subplot(1,2,2)
sns.boxplot(y='montos_desembolsados',data=df,color='lightcoral')
plt.title('BOXPLOT montos_desembolsados')

plt.show()

```
